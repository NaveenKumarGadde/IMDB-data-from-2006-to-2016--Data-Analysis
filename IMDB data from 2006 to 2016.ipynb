# Classifying Movies Based Rating

```python
Data.columns
```

```python
Data['Rating'].unique()
```

```python
def rating(number):
    if number >=7:
        return 'Excellent'
    elif number>=6:
        return 'Good'
    else:
        return 'Average'
```

```python
Data['Cat_rating'] = Data['Rating'].apply(rating)
```

# Finding unique values from Genre

```python
Data.columns
```

```python
Data['Genre']
```

```python
l1 =[]
for i in Data['Genre']:
    l1.append(i.split(','))

l1
   
```

```python
one_d = []

for i in l1:
    for j in i:
        one_d.append(j)
```

```python
one_d
```

```python
uni_list = []
for i in one_d:
    if i not in uni_list:
        uni_list.append(i)
```
